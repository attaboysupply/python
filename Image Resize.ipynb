```python
import PIL
import os
from PIL import Image
```

```python
f = r'C://Users/Attab/OneDrive/Desktop/test'
```

```python
os.listdir(f)
```

```python
#for loop to save out the image in a new size
for file in os.listdir(f):
    f_img = f+"/"+file
    img = Image.open(f_img)
    img = img.resize((512,512))
    #img = img.crop((100,100,612,612))
    img.save(f_img)
```
